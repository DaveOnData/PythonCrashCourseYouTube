```python
# Using a lambda for data wrangling (e.g., feature engineering)
reseller_sales_wrangled = (reseller_sales
                              .assign(Profit = lambda df_: df_['SalesAmount'] - df_['TotalProductCost'],
                                      ProfitPct = lambda df_: df_['Profit'] / df_['SalesAmount'],
                                      SalesExtendedDiff = lambda df_: df_['SalesAmount'] - df_['ExtendedAmount'],
                                      OrderYear = lambda df_: df_['OrderDate'].dt.year,
                                      OrderMonth = lambda df_: df_['OrderDate'].dt.month,
                                      OrderDay = lambda df_: df_['OrderDate'].dt.day,
                                      OrderQuarter = lambda df_: df_['OrderDate'].dt.quarter,
                                      OrderMonthName = lambda df_: df_['OrderDate'].dt.month_name(),
                                      OrderDayName = lambda df_: df_['OrderDate'].dt.day_name())
                          )

reseller_sales_wrangled
```

```python
# Calculate order line stats for SalesAmount column
(inner_join
    .groupby('SalesOrderNumber')
    ['SalesAmount']
    .agg(['sum', 'min', 'max', 'mean'])
)
```

```python
# Calculate order line stats for SalesAmount column
(inner_join
    .groupby('SalesOrderNumber')
    ['SalesAmount']
    .agg(['sum', 'min', 'max', 'mean'])
).reset_index()
```

```python
# Calculate order line stats for SalesAmount and Profit columns
(inner_join
    .groupby('SalesOrderNumber')
    [['SalesAmount', 'Profit']]
    .agg(['sum', 'min', 'max', 'mean'])
    .reset_index()
)
```

```python
# Group by SalesOrderNumber and define aggregations
(inner_join
    .groupby('SalesOrderNumber')
    .agg(SalesAmountTotal = ('SalesAmount', 'sum'),
         SalesAmountMin = ('SalesAmount', 'min'),
         ProfitTotal = ('Profit', 'sum'))
    .reset_index()
)
```

```python
# Group by Product Category and then by SalesOrderNumber
(inner_join
    .groupby(['ProductCategory', 'SalesOrderNumber'])
    .agg(SalesAmountTotal = ('SalesAmount', 'sum'),
         ProfitTotal = ('Profit', 'sum'))
    .reset_index()
)
```
